**Условие 3: 3 задача**

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.<br>
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
numeric_df = df.select_dtypes(include=[np.number])  # Выбор только числовых столбцов
corr_matrix = numeric_df.corr()  # Вычисление корреляционной матрицы
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

**Анализ и наблюдения**<br>
Корреляционная матрица - это таблица, которая показывает степень связи между различными переменными в наборе данных. Она используется для изучения взаимосвязи и зависимостей между переменными. Каждый элемент матрицы представляет собой коэффициент корреляции между двумя переменными. Значения коэффициента корреляции могут варьироваться от -1 до 1, где -1 указывает на полную отрицательную корреляцию, 1 - на положительную корреляцию, а 0 - на отсутствие корреляции.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
# Создание тепловой карты с уменьшенным размером шрифта
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm', annot_kws={"size": 6})
plt.show()
```

**Анализ и наблюдения**<br>
Визуализация корреляционной матрицы в виде тепловой карты позволяет наглядно представить степень корреляции между переменными. Тепловая карта использует цветовую шкалу, где разные оттенки цвета представляют различные значения корреляции. Это помогает исследователям быстро и легко определить сильные и слабые связи между переменными. Тепловая карта также позволяет выявить скрытые паттерны и структуры в данных.
Использование визуализации в виде тепловой карты помогает улучшить понимание данных и облегчает принятие решений на основе анализа корреляций. Это важный инструмент для исследования данных и может быть полезным при анализе экономических, финансовых, медицинских и других типов данных, где взаимосвязи между переменными играют важную роль.
